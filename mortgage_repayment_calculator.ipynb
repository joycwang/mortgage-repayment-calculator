## MORTGATE REPAYMENT CALCULATOR

# Overview:

This tool helps mortgage borrowers estimate the repayments on a home loan, based on their term and loan amount, and a fixed interest rate. It also examines the proportion of payments made in interest and principle over time. 

By plugging in the parameters such as purchase_price, loan to value ratio (LTV), interest rate(r) and term of the loan (n), individuals can work out their loan amount, monthly repayments and outstanding loan amount.

# Gather data on the home purchase price and loan details


Suppose your new home cost 1,000,000 AUD. You have saved 300,000 AUD for a deposit. You require another 700,000 AUD from a Bank to complete the purchase. Your Bank quotes the following Mortgage terms:

- Initial Loan Amount shall not exceed 70% of Purchase Price.
- Fixed interest rate: 5.5% p.a. (compounded monthly)
- Initial Term with fixed interest rate is 30 years
- Outstanding Loan Amount after 30 Years is zero
- Loan payments (interest & principle) shall be made monthly in arrears


```python
# Set the parameter values based on the above information
purchase_price = 1000000
LTV = 0.7
loan = home_price * LTV 
m = 12
n = 30
r = 0.05
```

# Calculate Monthly Repayment including interest and principle

```python
# rate = monthly interest rate 
# nper = number of monthly payments  
# pv = loan amount
# fv = Outstanding loan after 30yrs
cf = npf.pmt(rate = r/m, nper = n*m, pv = loan, fv = 0)
cf
```

# Calculate Principle repaid each month

```python
principle = npf.ppmt(rate = r/m, per = range(1, n*m + 1), nper = n*m, pv = loan, fv = 0)
principle
```

# Calculate Interest paid each month

```python
interest = npf.ipmt(rate = r/m, per = range(1, n*m + 1), nper = n*m, pv = loan, fv = 0)
interest
```

# Check if principle and interest add up to monthly repayment

```python
rem_loan = npf.fv(rate = r/m, nper = range(n*m + 1), pmt = cf, pv = loan)
rem_loan
```
